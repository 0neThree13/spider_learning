```python
import requests
from lxml import etree

headers = {
    'User-Agent':'Mozilla/5.0 (Linux; Android 6.0; Nexus 5 Build/MRA58N) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/134.0.0.0 Mobile Safari/537.36',
    }


session = requests.Session()
#session 是 requests库 提供的一个 会话管理对象，用于 在多个请求之间保持状态，比如 Cookie、Headers、Session ID 等。
# 通常，每次用 requests.get() 或 requests.post() 时，都是 一个新的 HTTP 连接，不保留登录状态。
# 而 使用 Session，可以在多个请求之间共享 Cookie，适用于：
# 模拟登录（登录后访问个人页面）
# 保持会话（避免重复登录，提升效率）✅
# 减少重复请求（复用 TCP 连接，加快访问速度）


#获取验证码
code_url='https://www.gushiwen.cn/user/login.aspx?from=http://www.gushiwen.cn/user/collect.aspx'

response = session.get(code_url, headers=headers)
tree = etree.HTML(response.text)


img_url='https://www.gushiwen.cn/' + tree.xpath('//*[@id="imgCode"]/@src')[0]
print(img_url)
img = session.get(img_url, headers=headers).content



with open("verify.jpg", "wb") as f:
    f.write(img)
```

### VIEWSTATE、VIEWSTATEGENERATOR
__VIEWSTATE 和 __VIEWSTATEGENERATOR 是 ASP.NET Web Forms 机制中的 页面状态管理参数。

__VIEWSTATE（视图状态）
###### •存储页面的一些状态信息，例如表单数据、用户交互数据等。
###### •服务器使用它来 维持页面状态，防止表单重复提交等问题。

__VIEWSTATEGENERATOR
###### •用于 唯一标识页面实例，确保 __VIEWSTATE 不会被伪造。
###### •通常是 ASP.NET 自动生成的一个标识码，验证请求是否合法。

直接提交表单数据是无效的，必须带上 最新的 __VIEWSTATE 和 __VIEWSTATEGENERATOR，否则：
###### •可能会 返回 500 服务器错误
###### •可能会 提示表单无效
###### •可能会 导致验证码一直不通过
