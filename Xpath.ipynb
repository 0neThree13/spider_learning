
# Xpath解析
###### XPath（XML Path Language）是一种用于在 XML 和 HTML 文档中定位元素的查询语言，常用于 网页爬取（Web Scraping） 和 XML 数据处理。
# 1.58同城二手房
######  58同城已经改用AJAX动态加载，未能从html中爬取数据

```python
from curl_cffi import requests
from lxml import etree

headers = {
    'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/91.0.4472.124 Safari/537.36',
}
url='https://m.58.com/sh/ershoufang/'

response=requests.get(url,headers=headers)
print(response.text)


tree=etree.HTML(response.text)
fp=open('58houses.txt','w',encoding='utf-8')

list = tree.xpath('//ul[@class="list"]/li')

for li in list:
    title=li.xpath('./a/div[2]/div[1]/text()')[0]
    fp.write(title+'\n')

```
